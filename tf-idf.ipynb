```python
import pandas as pd
```

```python
num = pd.read_csv('data/Trigger_Ave.csv')
```

```python
text = pd.read_csv('data/Trigger_Other.csv')
```

```python
text['corpus'] = [" ".join([str(a),str(b),str(c),str(d),str(e),str(f)]).lower().replace(',','').replace('.','').replace('"','').replace('?','').replace('   ',' ').replace('  ',' ').replace(' nan ',' ') for a,b,c,d,e,f in zip(text['t_q6'],text['t_q7'],text['t_q8'],text['t_q9'],text['t_q10'],text['t_q11'])]
```

```python
text['onegrams'] = [str(c).split(' ') for c in text['corpus']]
```

```python
chiefdom_list = list(text.groupby(['Chiefdom']).sum().index)
```

```python
chiefdoms = pd.DataFrame(columns=['chiefdom','corpus','onegrams'])
```

```python
for c in chiefdom_list:
    corpus = ' '.join(list(text[text['Chiefdom']==c]['corpus']))
    onegrams = str(corpus).split(' ')
    chiefdoms = chiefdoms.append({'chiefdom':c, 'corpus':corpus,'onegrams':onegrams},ignore_index=True)
```

```python
full_corpus = ' '.join(list(text['corpus']))
full_onegrams = str(full_corpus).split(' ')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(list(chiefdoms['corpus']))
```

```python
tf = pd.DataFrame(X.toarray(),columns=vectorizer.get_feature_names())
```

```python
tf
```

```python
import statistics as stat
```

```python
summary_stats = pd.DataFrame()
summary_stats['sum'] = tf.sum()
summary_stats['mean'] = tf.mean()
summary_stats['var'] = [stat.variance(tf[w]) for w in list(tf.columns)]
summary_stats = summary_stats.sort_values(by=['sum'],ascending=False)
```

```python
summary_stats = summary_stats.sort_values(by=['var'],ascending=False)
```
